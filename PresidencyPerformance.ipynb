```python
# Import Dependencies
import yfinance as yf
import pandas as pd
import datetime

import matplotlib.pyplot as plt
import seaborn as sns
```

**Ticker : Ticker Name**

"DX-Y.NYB" : **US Dollar**

"^GSPC" : **SP 500**

"^DJI" : **Dow Jones Industrial Average**

"^IXIC" : **NASDAQ**

"^RUT" : **Russell 2000**

```python
# Create a list of tickers to analyze
ticker_list = ["DX-Y.NYB", "^GSPC", "^DJI", "^IXIC", "^RUT"]

# Initialize an empty dataframe to hold the data
df = pd.DataFrame()

# iterate over each ticker in the list
for ticker in ticker_list:
    
    # Download the data for the current ticker from Yahoo Finance
    ticker_data = yf.download(ticker)

    # Add a column to the DataFrame to indicate which ticker the data belongs to
    ticker_data["Ticker"] = ticker
    
    # Add the data for the current ticker to the DataFrame
    df[ticker] = ticker_data["Adj Close"]

# Display the last few rows of the DataFrame
df.tail()
```

```python
# show the raw plot
df.plot();
```

List of US Presidents from 1971 to the current date (2023) in chronological order:

Richard Nixon **(1969-1974)**

Gerald Ford **(1974-1977)**

Jimmy Carter **(1977-1981)**

Ronald Reagan **(1981-1989)**

George H.W. Bush **(1989-1993)**

Bill Clinton **(1993-2001)**

George W. Bush **(2001-2009)**

Barack Obama **(2009-2017)**

Donald Trump **(2017-2021)**

Joe Biden **(2021-present)**

```python
# Define date ranges for each presidency
presidencies = {
    "Biden": ("2021-01-20", "2025-01-19"),
    "Trump": ("2017-01-20", "2021-01-19"),
    "Obama": ("2009-01-20", "2017-01-19"),
    "Bush Jr.": ("2001-01-20", "2009-01-19"),
    "Clinton": ("1993-01-20", "2001-01-19"),
    "Bush":("1989-01-20", "1993-01-19"),
    "Reagan":("1981-01-20", "1989-01-19"),
    "Carter":("1977-01-20", "1981-01-19"),
    "Ford":("1974-01-20", "1977-01-19"),
    "Nixon":("1969-01-20", "1974-01-19"),
}

# Create a new DataFrame for each presidency
for presidency, dates in presidencies.items():
    start_date, end_date = dates
    
    # Slice the original DataFrame for each presidency's dates
    df_presidency = df.loc[start_date:end_date]
    
    # Calculate the correlation matrix between the stocks
    correlation = df_presidency.corr()    
    
    # Create a heatmap of the correlation matrix
    sns.heatmap(correlation, cmap="coolwarm", annot=True)




    
    # Print the DataFrame for each respected presidency
    print(f"{presidency} Presidency:\n{df_presidency}\n")
    
    # Print the descriptive statistics for each respected presidency
    print(f"{presidency} Presidency Statistics:\n{df_presidency.describe()}\n")
    
    # Display the correlation matrix
    print(f"{presidency} Presidency Correlation Matrix:")
    print(correlation)    
    
    # Set the title of the plot
    plt.title(f" {presidency} Presidency Correlation Heatmap")
    # Show the plot
    plt.show()
    
    print("---------------------------------------------------------------------------------")
    print()
```

```python
# Create separate dataframes for each presidency

# Create an empty dictionary to store the separate dataframes for each presidency
president_dfs = {}

# Loop through each presidency and its corresponding dates in the presidencies dictionary
for president, dates in presidencies.items():
    
    # Unpack the start and end dates for the current presidency
    start_date, end_date = dates
    
    # Slice the original DataFrame for each presidents dates
    # Create a boolean mask for the rows in the DataFrame that fall within the current presidency's dates
    mask = (df.index >= start_date) & (df.index < end_date)
    
    # Slice the DataFrame using the mask and create a copy to avoid modifying the original DataFrame
    president_df = df.loc[mask].copy()
    
    # Add the new DataFrame to the dictionary with the current presidency as the key
    president_dfs[president] = president_df

# Plot the data

 # Create a figure with subplots for each presidency
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 8))

# Loop through each presidency and its corresponding DataFrame in the president_dfs dictionary
for i, (president, president_df) in enumerate(president_dfs.items()):
    
    # Calculate the row index for the current subplot
    row = i // 5
    
    # Calculate the column index for the current subplot
    col = i % 5
    
    # Get the current subplot
    ax = axes[row, col]
    
    # Set the title of the subplot to the current presidency
    ax.set_title(president)
    
    # Plot the data for the current presidency on the current subplot, with a legend
    president_df.plot(ax=ax,)# legend=None)
    
    # Set the x-axis label to an empty string
    ax.set_xlabel("")
    
    # Set the y-axis label to "Adj Close"
    ax.set_ylabel("Adj Close")
    
    # Add a grid to the subplot with an alpha value of 0.5
    ax.grid(alpha=0.5)

# Adjust the subplots to fit within the figure
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a separate plot for each ticker, and then loops over each presidency to plot the data for that presidency. 
# Highlight each presidency in a different color

# Loop over each ticker
for ticker in ticker_list:
    
    # Create a figure and set the size
    fig, ax = plt.subplots(figsize=(12, 6))

    # Loop over each presidency
    for presidency, dates in presidencies.items():

        # Select the data for the presidency
        presidency_data = df.loc[dates[0]:dates[1], ticker]

        # Plot the data
        ax.plot(presidency_data, label=presidency)

    # Set the title and axis labels
    ax.set_title(ticker)
    ax.set_xlabel("Date")
    ax.set_ylabel("Adjusted Close Price")

    # Add a legend and show the plot
    ax.legend()
    plt.show()
```

```python
# Loop over each ticker
for ticker in ticker_list:
    
    # Create a new figure with subplots for each presidency
    fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 8))
    
    # Download the data for the current ticker from Yahoo Finance
    ticker_data = yf.download(ticker)
    
    # Loop through each presidency and its corresponding DataFrame in the president_dfs dictionary
    for i, (president, president_df) in enumerate(president_dfs.items()):
        
        # Calculate the row index for the current subplot
        row = i // 5
        
        # Calculate the column index for the current subplot
        col = i % 5
        
        # Get the current subplot
        ax = axes[row, col]
        
        # Set the title of the subplot to the current presidency
        ax.set_title(president, fontsize=20)
        
        # Set the x-axis label to an empty string
        ax.set_xlabel("")
        
        # Set the y-axis label to "Adj Close"
        ax.set_ylabel("Adj Close")
        
        # Add a grid to the subplot with an alpha value of 0.5
        ax.grid(alpha=0.5)
        
        # Plot the data for the current presidency on the current subplot
        start_date, end_date = presidencies[president]
        president_data = ticker_data.loc[start_date:end_date]
        color = sns.color_palette("bright")[i]
        president_data["Adj Close"].plot(ax=ax, label=president, color=color)
    
    # Set the title of the figure to the current ticker
    fig.suptitle(ticker)
    
    # Add a legend to the figure
    # handles, labels = ax.get_legend_handles_labels()
    # fig.legend(handles, labels, loc="upper left")
    
    # Adjust the subplots to fit within the figure
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```

```python
# create a bar chart to show the performance of each stock during each presidency

# Create an empty dataframe to hold the mean adjusted close price for each stock during each presidency
presidency_means = pd.DataFrame(columns=presidencies.keys(), index=df.columns)

# Loop through each presidency and its corresponding DataFrame in the president_dfs dictionary
for president, president_df in president_dfs.items():
    
    # Loop through each stock in the DataFrame
    for stock in president_df.columns:
        
        # Calculate the mean adjusted close price for the current stock during the current presidency
        mean = president_df[stock].mean()
        
        # Add the mean to the presidency_means dataframe at the current stock and presidency
        presidency_means.loc[stock, president] = mean

# Transpose the dataframe to make it easier to plot
presidency_means = presidency_means.T

# Create a bar chart of the mean adjusted close price for each stock during each presidency
presidency_means.plot(kind="bar", figsize=(10, 5))

# Set the title of the plot
plt.title("Mean Adjusted Close Price by Presidency")

# Set the x-axis label to the name of each presidency
plt.xlabel("Presidency")

# Set the y-axis label to "Mean Adjusted Close Price"
plt.ylabel("Mean Adjusted Close Price")

# Show the plot
plt.show()
```

```python
# create a correlation matrix heatmap to show the performance of each stock during each presidency

# Create a figure with subplots for each presidency
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 8))

# Loop through each presidency and its corresponding dates in the presidencies dictionary
for i, (president, dates) in enumerate(presidencies.items()):
    
    # Unpack the start and end dates for the current presidency
    start_date, end_date = dates
    
    # Slice the original DataFrame for each presidency's dates
    df_presidency = df.loc[start_date:end_date]
    
    # Calculate the correlation matrix between the stocks
    correlation = df_presidency.corr()    
    
    # Get the current subplot
    row = i // 5
    col = i % 5
    ax = axes[row, col]
    
    # Create a heatmap of the correlation matrix in the current subplot
    sns.heatmap(correlation, cmap="coolwarm", annot=True, ax=ax)
    
    # Set the title of the current subplot to the current presidency
    ax.set_title(president)
    
    # Set the x and y axis labels to empty strings to save space
    ax.set_xlabel("")
    ax.set_ylabel("")

# Adjust the layout of the subplots to avoid overlapping labels
plt.tight_layout()

# Show the plot
plt.show()
```
